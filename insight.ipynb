```python
#linerRegression and predictions
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
Suicide_EPI_CLEAN1=Suicide_EPI_CLEAN.drop(['country','year','database\SO2'], axis=1)

Suicide_EPI_CLEAN1_final=Suicide_EPI_CLEAN1.loc[Suicide_EPI_CLEAN1['database\\UWD'].isnull()==False]
for column in Suicide_EPI_CLEAN1_final:
    Suicide_EPI_CLEAN1_final[column].fillna(Suicide_EPI_CLEAN1_final[column].mean(),inplace=True)
```

```python
y_suicide=Suicide_EPI_CLEAN1_final[Suicide_EPI_CLEAN1_final.columns[0]]
X_suicide=Suicide_EPI_CLEAN1_final[Suicide_EPI_CLEAN1_final.columns[1:]]
lm = linear_model.LinearRegression()
model = lm.fit(X_suicide,y_suicide)
predictions = model.predict(X_suicide)
plot.title('Predictions')
plot.xlabel("y_suicide")
plot.ylabel("predictions")
plot.scatter(y_suicide, predictions, color='red')
plot.show()
residual = predictions - y_suicide
plot.title('Predictions residual')
plot.xlabel("predictions")
plot.ylabel("residual")
plot.scatter(predictions, residual, color='red')
plot.show()
```

```python
print( "score:",lm.score(X_suicide,y_suicide),'\n',"coef:",lm.coef_,'\n',"intercept:", lm.intercept_)
```
